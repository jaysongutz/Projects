```python
import pandas as pd
import numpy as np
import math as m
from matplotlib import pyplot as py
```

```python
#1. Using Google Sheets, make an appropriate graph to visualize the relationship between
# the number of Instagram followers and the daily average of phone screen time last week
# (in mins)?
df = pd.read_csv('./Student Survey - Sheet1.csv')

df.sort_values('#Instagram followers', inplace=True)

df = df.reset_index()

instafollow = list(df['#Instagram followers'])

phonescreen = list(df['Phone screen time (mins)'])

py.scatter(instafollow, phonescreen)
py.title('Daily Phone Screen Time Based on Instagram Following')
py.xlabel('# Instagram followers')
py.ylabel('Avg. Daily Phone Screen Time (mins)')
```

```python
#9. Calculate the regression line between Student Height and Mother Height. Use Mother
# Height as the predictor (i.e. X) and Student Height as the prediction (i.e. Y). What is the
# intercept ?
x = list(df['Mother Height'])
y = list(df['Student Height'])

#calculate mean of x and y
sumx = 0.0
sumy = 0.0

for i in range(len(x)):
    sumx += x[i]
for i in range(len(y)):
    sumy += y[i]

meanx = sumx/len(x)
meany = sumy/len(y)

#calculate denom
ssx = 0.0
spxy = 0.0
for i in range(len(x)):
    ssx += (x[i] - meanx) * (x[i] - meanx)
    spxy += (x[i] - meanx) * (y[i] - meany)

slope = spxy /ssx
print(slope)

intercept = meany - (slope * meanx)
print(intercept)
```
